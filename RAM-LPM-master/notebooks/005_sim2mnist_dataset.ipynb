```python
from torchvision import datasets, transforms
from torchex.data import transforms as extransforms
import matplotlib.pyplot as plt
```

```python
data_root = '~/.mnist'
train_dataset = datasets.MNIST(data_root, train=True, download=True,
                             transform=transforms.Compose([
                                    extransforms.RandomResize(28, 68, True),
                                    transforms.RandomRotation((-180, 180)),
                                    extransforms.PadRandomSift(96, 96),
                                    transforms.ToTensor(), 
                                   ]))
```

```python
def tensor2img(data):
    fig = plt.figure()
    ax = fig.add_subplot(121)
    x = data.data.numpy()[0, :, :]
    ax.imshow(x)
    plt.show()
```

```python
tensor2img(train_dataset[0][0])
```
